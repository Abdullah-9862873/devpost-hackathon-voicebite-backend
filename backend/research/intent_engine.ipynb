```python
import google.generativeai as genai
```

```python
import os
```

```python
from dotenv import load_dotenv
# Load from backend root
load_dotenv('../../.env')
```

```python
for m in genai.list_models():
    if 'generateContent' in m.supported_generation_methods:
        print(m.name)
```

```python
# Setup
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))
model = genai.GenerativeModel('gemini-flash-latest')
print("✅ Ready!")
```

```python
PROMPT_CONTEXT = """
You are the VoiceBite AI Intent Engine. Convert speech to JSON.
ACTIONS: 
- GET_CATEGORY (category: pizza|pasta|desserts)
- ADD_TO_CART (name: exact item name)
- SEARCH (query: text)
- NAVIGATE (page: home|cart)
Example: "Show me pizzas" -> {"action": "GET_CATEGORY", "payload": {"category": "pizza"}}
"""
def get_intent(text):
    response = model.generate_content(f"{PROMPT_CONTEXT}\n\nUser: {text}")
    return response.text.strip()
```

```python
# Testing 
print(get_intent("Open Beverages"))
```
